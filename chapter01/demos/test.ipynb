# Demo

This Notebook mainly tests whether the interactive environment is working.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, HTML

%matplotlib widget  

# Create a figure with fixed size
fig, ax = plt.subplots(figsize=(6, 3))

# Remove scrollbars using CSS
display(HTML("<style>div.jp-OutputArea-output { overflow: hidden; }</style>"))

def interactive_plot(frequency, amplitude):
    x = np.linspace(0, 10, 1000)
    y = amplitude * np.sin(frequency * x)

    ax.clear()  # Prevents multiple plots from stacking
    ax.plot(x, y, label=f"Sine Wave: {frequency} Hz, {amplitude} Amplitude")
    ax.set_xlabel("Time")
    ax.set_ylabel("Amplitude")
    ax.grid(True)

    # Prevent the extra "Figure" print
    fig.canvas.draw_idle()  

# Disable toolbar (prevents zoom-based scrollbars)
fig.canvas.toolbar_visible = False  
fig.canvas.header_visible = False  
fig.canvas.footer_visible = False  

# Interactive controls
freq_slider = widgets.FloatSlider(min=0.1, max=10.0, step=0.1, value=1.0, description="Frequency")
amp_slider = widgets.FloatSlider(min=0.1, max=2.0, step=0.1, value=1.0, description="Amplitude")
ui = widgets.VBox([freq_slider, amp_slider])

# Connect sliders to function
output = widgets.interactive_output(interactive_plot, {'frequency': freq_slider, 'amplitude': amp_slider})

# Display only the controls (no unwanted "Figure" text)
display(ui, output)
```
